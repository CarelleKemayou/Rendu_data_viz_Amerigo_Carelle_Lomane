```python
def get_100(liste:list):# verifies les pourcentages 
        liste=[float(x) for x in liste]
        sum=0
        for n in liste:
            sum+=n
        if sum !=100:
            percentage=100-sum
            liste.append(percentage)

        return liste
```

```python




import os 
import json

def export_all_columns(file_path,output_folder):#converteur json -->js
    df = pd.read_json(file_path)
    
    # On crée le nom du fichier de sortie
    filename_only = os.path.basename(file_path).replace(".json","")#Extrait  le nom du fichier du path 
    output_file = os.path.join(output_folder, f"variables_{filename_only}.js")#Genere un path pour le nouveau fichier
    
    os.makedirs(output_folder, exist_ok=True)# cree le fichier si il n'existe pas 


    with open(output_file, "w") as f:
        for col_name in df.columns:
            liste_valeurs = df[col_name].tolist()#Convertis colonnes en list 
           
            liste_valeurs = [ None  if pd.isna(v) else v for v in liste_valeurs]#Convertit les  None en 0


            if "percent" in col_name:# corrige le % pour les colonnes 
                liste_valeurs = get_100(liste_valeurs)
                print(f"Colonne '{col_name}' corrigée.")
            if "category" in col_name:# corrige le % pour les colonnes 
                liste_valeurs.append('Autres')
                print(f"Colonne '{col_name}' corrigée.")

            if "date" in col_name :
                liste_valeurs=[str(v).replace("Timestamp(","").replace(" 00:00:00')","'").replace("'", "")# Retire les timestamps et heures des dates
    .replace(" 00:00:00", "") for v in liste_valeurs] 

            # Écriture dans le fichier JS
            col_name_js = col_name.replace(" ", "_").replace("+", "plus")# corrige les nom des listes
            f.write(f"const {col_name_js} = {json.dumps(liste_valeurs)};\n")#Corrige automatiquement les None en null
            print(f"Colonne '{col_name}' exportée.")

```

```python


import os 

def convert(folder_path,output_folder):
    for filename in os.listdir(folder_path):#bouble sur chauqe fichier  dans le dossier
        file_path = os.path.join(folder_path, filename)#garde le nom du fichier et cree un chemin 
        if os.path.isfile(file_path):#verifie si c'est un fichier
            print(f"Traitement du fichier : {filename}")
            export_all_columns(file_path,output_folder)# l'exporteu

```
